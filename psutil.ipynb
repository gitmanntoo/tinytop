```python
for f in sorted(dir(platform)):
    if f.startswith("_"):
        continue

    fn = getattr(platform, f)
    if callable(fn):
        try:
            result = fn()
        except Exception as e:
            result = f"Error: {e}"
        print(f"{f}(): {result}")
```
